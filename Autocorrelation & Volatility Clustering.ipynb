```python
# Downloading the Apple Share Price Data from Yahoo Finance

import yfinance as yf

df  = yf.download (
    "AAPL",
    period = "1y",
    interval = "1wk"
)
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df
```

```python
# log-returns
returns = np.log(df['Close']).diff().dropna()
returns
```

```python
returns.plot()
plt.title('Weekly Returns of Apple Stock')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid()
plt.show()
```

```python
#Resampling to Weekly and Monthly Returns

weekly_returns = returns.resample('W').sum()
weekly_returns.plot()
plt.title('Weekly Returns of Apple Stock')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid()
plt.show()

monthly_returns = returns.resample('M').sum()
monthly_returns.plot()
plt.title('Monthly Returns of Apple Stock')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid()
plt.show()
```

```python
# Autocorrelation Analysis
returns.squeeze().autocorr(lag=1)
from statsmodels.tsa.stattools import acf

nlags = 40
acf_values = acf(returns, nlags=40, fft=True)

conf = 1.96 / np.sqrt(len(returns))

acf_returns = acf(returns, nlags=nlags, fft=True)
acf_sq_returns = acf(returns**2, nlags=nlags, fft=True)

plt.figure(figsize=(10,4))
plt.stem(range(nlags+1), acf_returns, )
plt.axhline(conf, linestyle='--', color='red')
plt.axhline(-conf, linestyle='--', color='red')
plt.axhline(0, color='black')
plt.title("ACF of Returns")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.show()

plt.figure(figsize=(10,4))
plt.stem(range(nlags+1), acf_sq_returns, )
plt.axhline(conf, linestyle='--', color='red')
plt.axhline(-conf, linestyle='--', color='red')
plt.axhline(0, color='black')
plt.title("ACF of Squared Returns (Volatility)")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.show()
```
