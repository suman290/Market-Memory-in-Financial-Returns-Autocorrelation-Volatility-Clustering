```python
# Downloading the Apple Share Price Data from Yahoo Finance

import yfinance as yf

df  = yf.download (
    "AAPL",
    period = "1y",
    interval = "1wk"
)
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df
```

```python
# log-returns
returns = np.log(df['Close']).diff().dropna()
returns
```

```python
returns.plot()
plt.title('Weekly Returns of Apple Stock')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid()
plt.show()
```

```python
#Resampling to Weekly and Monthly Returns

weekly_returns = returns.resample('W').sum()
weekly_returns.plot()
plt.title('Weekly Returns of Apple Stock')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid()
plt.show()

monthly_returns = returns.resample('ME').sum()
monthly_returns.plot()
plt.title('Monthly Returns of Apple Stock')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid()
plt.show()
```

```python
# Autocorrelation Analysis
returns.squeeze().autocorr(lag=1)
from statsmodels.tsa.stattools import acf

nlags = 40
acf_values = acf(returns, nlags=40, fft=True)

conf = 1.96 / np.sqrt(len(returns))

acf_returns = acf(returns, nlags=nlags, fft=True)
acf_sq_returns = acf(returns**2, nlags=nlags, fft=True)

plt.figure(figsize=(10,4))
plt.stem(range(nlags+1), acf_returns, )
plt.axhline(conf, linestyle='--', color='red')
plt.axhline(-conf, linestyle='--', color='red')
plt.axhline(0, color='black')
plt.title("ACF of Returns")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.show()

plt.figure(figsize=(10,4))
plt.stem(range(nlags+1), acf_sq_returns, )
plt.axhline(conf, linestyle='--', color='red')
plt.axhline(-conf, linestyle='--', color='red')
plt.axhline(0, color='black')
plt.title("ACF of Squared Returns (Volatility)")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.show()
```

```python
shuffled = returns.sample(frac=1, replace=False).values

shuffled = np.random.permutation(returns.values)
```

```python
block_size = 10
blocks = []

for i in range(0, len(returns), block_size):
    blocks.append(returns.iloc[i:i+block_size])

np.random.shuffle(blocks)
block_shuffled = pd.concat(blocks)

```

```python
shuffled = np.random.permutation(returns.values.flatten())
shuffled = pd.Series(shuffled, index=returns.index)

block_shuffled = np.random.permutation(returns.values.flatten())
block_shuffled = pd.Series(block_shuffled, index=returns.index)
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

lags = [10, 20, 30]

lb_real = acorr_ljungbox(returns**2, lags=lags, return_df=True)
lb_shuffle = acorr_ljungbox(shuffled**2, lags=lags, return_df=True)
lb_block = acorr_ljungbox(block_shuffled**2, lags=lags, return_df=True)

print("\nLjungBox p-values (Squared Returns)\n")

for lag in lags:
    print(f"Lag {lag}:")
    print(f"  Real        p-value: {lb_real.loc[lag, 'lb_pvalue']:.6f}")
    print(f"  Shuffled    p-value: {lb_shuffle.loc[lag, 'lb_pvalue']:.6f}")
    print(f"  Block       p-value: {lb_block.loc[lag, 'lb_pvalue']:.6f}")
    print("-"*40)
```

```python
from statsmodels.tsa.stattools import acf
import numpy as np

nlags = 40

acf_real = acf(returns**2, nlags=nlags)
acf_shuffle = acf(shuffled**2, nlags=nlags)
acf_block = acf(block_shuffled**2, nlags=nlags)

mean_real = np.mean(acf_real[1:11])
mean_shuffle = np.mean(acf_shuffle[1:11])
mean_block = np.mean(acf_block[1:11])

print("\nAverage Squared-Return ACF (lags 110):\n")
print(f"Real:      {mean_real:.6f}")
print(f"Shuffled:  {mean_shuffle:.6f}")
print(f"Block:     {mean_block:.6f}")
```

```python
import matplotlib.pyplot as plt

lags = range(nlags + 1)

plt.figure(figsize=(10,5))
plt.plot(lags, acf_real, label="Real")
plt.plot(lags, acf_shuffle, label="Shuffled")
plt.plot(lags, acf_block, label="Block-Shuffled")

plt.axhline(0)
plt.title("Squared Return ACF Comparison")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.legend()
plt.show()
```

conclusion
While the marginal distribution of returns remains unchanged under permutation, the volatility autocorrelation structure collapses under both full and block shuffling. This indicates that observed volatility persistence arises from genuine temporal dependence rather than distributional characteristics alone.
